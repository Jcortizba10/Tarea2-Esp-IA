```python
plt.style.use('classic')
imagen = Image.open("./imagen/imag.jpeg")
# Mostramos la imagen usando matplotlib
plt.imshow(imagen)
# Mostramos la imagen usando matplotlib
plt.imshow(imagen)
#plt.axis('off')  # Para no mostrar los ejes
#plt.show()
```

```python
imagen_gr = imagen.convert('LA')
imagen_mat = np.array(list(imagen_gr.getdata(band=0)),float)
imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
```

```python
U,D,V = np.linalg.svd(imagen_mat)
```

```python
print(imagen_mat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

```python
imagen_recons = np.matrix(U[:, :1])* np.diag(D[:1])*np.matrix(V[:1,1:])
plt.imshow(imagen_recons, cmap='gray')
```

```python
i=5 #valor singular
imagen_recons = np.matrix(U[:, :i])* np.diag(D[:i])*np.matrix(V[:i,:])
plt.imshow(imagen_recons, cmap='gray')
title= 'valores singulares = %s' % i
plt.title(title)
plt.show()
```

```python
i=15
imagen_recons = np.matrix(U[:, :i])* np.diag(D[:i])*np.matrix(V[:i,:])
plt.imshow(imagen_recons, cmap='gray')
title= 'valores singulares = %s' % i
plt.title(title)
plt.show()
```

```python
i=60
imagen_recons = np.matrix(U[:, :i])* np.diag(D[:i])*np.matrix(V[:i,:])
plt.imshow(imagen_recons, cmap='gray')
title= 'valores singulares = %s' % i
plt.title(title)
plt.show()
```

```python
# Verificar si la imagen se ha le铆do correctamente
if imagen is None:
    raise ValueError(f"Error al cargar la imagen desde la ruta: {image_path}")

# Obtener la forma de la imagen
m, n = imagen.shape[:2]
print(f"Filas: {m}")
print(f"Columnas: {n}")

# Calcular la cantidad de bytes de la imagen original
cant_bytes_original = m * n * 3
print('Cantidad de bytes a utilizar para crear la imagen:', cant_bytes_original)

# Convertir la imagen de BGR a RGB
image_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Mostrar la imagen usando matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Ocultar los ejes
plt.show()
```
