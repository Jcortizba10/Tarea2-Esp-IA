Vamos a tomar como ejemplo el An√°lisis de Componentes Principales (PCA) para ilustrar c√≥mo se aplican los conceptos de √°lgebra lineal en el aprendizaje autom√°tico.

An√°lisis de Componentes Principales (PCA)
En palabras sencillas el PCA es una t√©cnica del an√°lisis multivariado (estad√≠stica muy avanzada) que combina conceptos de √°lgebra lineal y estadistica y la finalidad de este m√©todo es REDUCIR, ojo!, no eliminar sino reducir la misma informaci√≥n de todas nuestras variables (dimensiones) a unas nuevas variables que les vamos a llamar componentes principales, y estas variables van a 1. explicar la misma informaci√≥n que las variables m√°s fuertes o con m√°s variabilidad 2. contener adicionalmente la mayor informaci√≥n posible de las otras dimensiones que no arrojen tanta informaci√≥n o no haya tanta correlaci√≥n.

PCA se basa en varios conceptos de √°lgebra lineal, tales como:

Matrices y Vectores: Los datos de entrada se representan como una matriz, donde cada fila es una observaci√≥n y cada columna es una caracter√≠stica.
Media y Matriz de Covarianza: Se calcula la media de los datos y se centra la matriz de datos restando la media. Luego, se calcula la matriz de covarianza de los datos centrados.
Autovalores y Autovectores: Se calculan los autovalores y autovectores de la matriz de covarianza. Los autovectores (tambi√©n llamados componentes principales) determinan la nueva base para los datos transformados.
Transformaci√≥n de Datos: Los datos originales se proyectan sobre los autovectores para obtener los componentes principales.

```python
def graficarVectores(vecs, cols, alpha = 1):
    
    plt.axvline(x = 0, color = "grey", zorder = 0)
    plt.axhline(y = 0, color = "grey", zorder = 0)
    
    for i in range(len(vecs)):
        x = np.concatenate([[0,0], vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles = 'xy', scale_units = 'xy',
                   scale = 1,
                   color = cols[i],
                   alpha = alpha
                  )
```
